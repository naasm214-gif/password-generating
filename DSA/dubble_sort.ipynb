```python
def dubble_sort(arr):
    n=len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
print(dubble_sort([64,34,25,22,11,90]))
```

```python

def sort_fun(sort):
    a=len(sort)
    for m in range(a):
        for n in range(0,a-m-1):
            if sort[n]>sort[n+1]:
                sort[n],sort[n+1]=sort[n+1],sort[n]
    return sort

print(sort_fun([1,5,6,7,3,2,5,7,9,3]))
```
